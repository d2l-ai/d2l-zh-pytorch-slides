常见的标准算术运算符（`+`、`-`、`*`、`/` 和 `**`）都可以被升级为逐元素运算

```python
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])
x + y, x - y, x * y, x / y, x**y  # **运算符是求幂运算
```

按逐元素方式应用更多的计算

```python
torch.exp(x)
```

```python
X = torch.arange(12, dtype=torch.float32).reshape((3, 4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
torch.cat((X, Y), dim=0), torch.cat((X, Y), dim=1)
```

```python
X == Y
```

```python
X.sum()
```

即使形状不同，我们仍然可以通过调用 *广播机制*（broadcasting mechanism） 来执行逐元素操作

```python
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
```

```python
a + b
```

可以用`[-1]` 选择最后一个元素，可以用`[1:3]` 选择第二个元素和第三个元素

```python
X[-1], X[1:3]
```

```python
X[1, 2] = 9
X
```

```python
X[0:2, :] = 12
X
```

```python
before = id(Y)
Y = Y + X
id(Y) == before
```

执行原地操作

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

```python
before = id(X)
X += Y
id(X) == before
```

```python
A = X.numpy()
B = torch.tensor(A)
type(A), type(B)
```
